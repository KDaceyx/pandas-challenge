```python
 # Dependencies and Setup
    
import pandas as pd
```

```python
# File to Load

source_file = "purchase_data.csv"
```

```python
# Read Purchasing File and store into Pandas data frame

source_data = pd.read_csv(source_file)
source_data
```

```python
# Display the total number of players

count_players = source_data['SN'].value_counts()
Total_Players = len(count_players)

# create a new dataframe for the summary report

reportDF = pd.DataFrame()
reportDF['Total Players'] = [Total_Players]
reportDF
```

```python
# Purchasing Analysis(total):

# Number of Unique Items
NumberItems = sum(source_data['Item Name'].value_counts())

# Average Purchase Price
AvgPrice = source_data['Price'].mean()

# Total Number of Purchases
TotalPurchases = len(source_data['Purchase ID'])

# Total Revenue
TotalRevenue = source_data['Price'].sum()
```

```python
# Display summary table

SummaryDF = pd.DataFrame({"Total Players": [Total_Players],
                          "Number of Unique Items": [NumberItems],
                          "Average Purchase Price": [AvgPrice],
                         "Number of Purchases": [TotalPurchases],
                          "Total Revenue": [TotalRevenue]})

# Format Columns

SummaryDF["Average Purchase Price"] = SummaryDF["Average Purchase Price"].map("${:,.2f}".format)
SummaryDF["Total Revenue"] = SummaryDF["Total Revenue"].map("${:,.2f}".format)


SummaryDF
```

```python
# Create DF for unique SNs only
UniqueSN = source_data[["SN","Gender"]]
UniqueSN = UniqueSN.drop_duplicates(subset = ["SN"])
UniqueSN
```

```python
#Gender Demographics
#Percentage and Count of Male Players

MaleCount = len(UniqueSN[UniqueSN['Gender'] == 'Male'])
MalePercent = (MaleCount / Total_Players) * 100

#Percentage and Count of Female Players

FemaleCount = len(UniqueSN[UniqueSN['Gender'] == 'Female'])
FemalePercent = (FemaleCount / Total_Players) * 100

#Percentage and Count of Other / Non-Disclosed

OtherCount = len(UniqueSN[UniqueSN['Gender'] == 'Other / Non-Disclosed'])
OtherPercent = (OtherCount / Total_Players) * 100

# Gender Summary

GenderSummary = pd.DataFrame({'Gender': ['Male', 'Female', 'Other / Non-Disclosed'],
                             'Total Count': [MaleCount, FemaleCount, OtherCount],
                             'Percentage of Players': [MalePercent, FemalePercent, OtherPercent]})
GenderSummary.set_index('Gender')
GenderSummary['Percentage of Players'] = GenderSummary['Percentage of Players'].map('{:.2f}%'.format)

GenderSummary
```

```python
# Purchasing Analysis (Gender)
# Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender

MaleDF = source_data[source_data.Gender == 'Male']
MalePurchaseCount = len(MaleDF)
AvgMalePurchase = MaleDF['Price'].mean()
MaleValueTotal = MaleDF['Price'].sum()
MalePerPerson = (MaleValueTotal / MaleCount)

FemaleDF = source_data[source_data.Gender == 'Female']
FemalePurchaseCount = len(FemaleDF)
AvgFemalePurchase = FemaleDF['Price'].mean()
FemaleValueTotal = FemaleDF['Price'].sum()
FemalePerPerson = (FemaleValueTotal / FemaleCount)

OtherDF = source_data[source_data.Gender == 'Other / Non-Disclosed']
OtherPurchaseCount = len(OtherDF)
AvgOtherPurchase = OtherDF['Price'].mean()
OtherValueTotal = OtherDF['Price'].sum()
OtherPerPerson = (OtherValueTotal / OtherCount)

# Create a summary data frame to hold the results
GenderPurchasesDF = pd.DataFrame({'Gender': ['Male', 'Female', 'Other / Non-Disclosed'],
                                 'Purchase Count': [MalePurchaseCount, FemalePurchaseCount, OtherPurchaseCount],
                                 'Average Purchase Price': [AvgMalePurchase, AvgFemalePurchase, AvgOtherPurchase],
                                 'Total Purchase Value': [MaleValueTotal, FemaleValueTotal, OtherValueTotal],
                                 'Avg Total Purchase Per Person': [MalePerPerson, FemalePerPerson, OtherPerPerson]})
GenderPurchasesDF.set_index('Gender')
GenderPurchasesDF['Average Purchase Price'] = GenderPurchasesDF['Average Purchase Price'].map('${:.2f}'.format)
GenderPurchasesDF['Total Purchase Value'] = GenderPurchasesDF['Total Purchase Value'].map('${:.2f}'.format)
GenderPurchasesDF['Avg Total Purchase Per Person'] = GenderPurchasesDF['Avg Total Purchase Per Person'].map('${:.2f}'.format)
GenderPurchasesDF
```

```python
# Top Spenders
# Identify top 5 spenders by purchase count
# group by SN
Top_Spenders = source_data.groupby(['SN'])
```

```python
# Make and format a new summary table
SpendersDF = pd.DataFrame({'Purchase Count':Top_Spenders['Price'].count(),
                          'Average Purchase Price': Top_Spenders['Price'].mean(),
                          'Total Purchases': Top_Spenders['Price'].sum()})
SpendersDF = SpendersDF.sort_values('Total Purchases', ascending=False)

SpendersDF['Average Purchase Price'] = SpendersDF['Average Purchase Price'].map('${:.2f}'.format)
SpendersDF['Total Purchases'] = SpendersDF['Total Purchases'].map('${:.2f}'.format)
SpendersDF.head()
```

```python
# Most Popular Items
# Retrieve the Item ID, Item Name, and Item Price columns
# Group by Item ID and Item Name
# Obtain purchase count, average item price, and total purchase value
# Format and display a preview of the summary data frame

PopularGroup = source_data.groupby(['Item ID', 'Item Name'])
PopularDF = pd.DataFrame({'Purchase Count': PopularGroup['Price'].count(),
                         'Item Price': PopularGroup['Price'].mean(),
                         'Total Purchase Value': PopularGroup['Price'].sum()})
PopularDF = PopularDF.sort_values('Purchase Count', ascending=False)
PopularDF['Item Price'] = PopularDF['Item Price'].map('${:.2f}'.format)
PopularDF['Total Purchase Value'] = PopularDF['Total Purchase Value'].map('${:.2f}'.format)
PopularDF.head()
```
