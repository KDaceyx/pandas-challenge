```python
 # Dependencies and Setup
    
import pandas as pd
```

```python
# File to Load

source_file = "purchase_data.csv"
```

```python
# Read Purchasing File and store into Pandas data frame

source_data = pd.read_csv(source_file)
source_data
```

```python
# Display the total number of players

count_players = source_data['SN'].value_counts()
Total_Players = len(count_players)

# create a new dataframe for the summary report

reportDF = pd.DataFrame()
reportDF['Total Players'] = [Total_Players]
reportDF
```

```python
# Purchasing Analysis(total):

# Number of Unique Items
NumberItems = sum(source_data['Item Name'].value_counts())

# Average Purchase Price
AvgPrice = source_data['Price'].mean()

# Total Number of Purchases
TotalPurchases = len(source_data['Purchase ID'])

# Total Revenue
TotalRevenue = source_data['Price'].sum()
```

```python
# Display summary table

SummaryDF = pd.DataFrame({"Total Players": [Total_Players],
                          "Number of Unique Items": [NumberItems],
                          "Average Purchase Price": [AvgPrice],
                         "Number of Purchases": [TotalPurchases],
                          "Total Revenue": [TotalRevenue]})

# Format Columns

SummaryDF["Average Purchase Price"] = SummaryDF["Average Purchase Price"].map("${:,.2f}".format)
SummaryDF["Total Revenue"] = SummaryDF["Total Revenue"].map("${:,.2f}".format)


SummaryDF
```

```python
# Create DF for unique SNs only
UniqueSN = source_data[["SN","Gender"]]
UniqueSN = UniqueSN.drop_duplicates(subset = ["SN"])
UniqueSN
```

```python
#Gender Demographics
#Percentage and Count of Male Players

MaleCount = len(UniqueSN[UniqueSN['Gender'] == 'Male'])
MalePercent = (MaleCount / Total_Players) * 100

#Percentage and Count of Female Players

FemaleCount = len(UniqueSN[UniqueSN['Gender'] == 'Female'])
FemalePercent = (FemaleCount / Total_Players) * 100

#Percentage and Count of Other / Non-Disclosed

OtherCount = len(UniqueSN[UniqueSN['Gender'] == 'Other / Non-Disclosed'])
OtherPercent = (OtherCount / Total_Players) * 100

# Gender Summary

GenderSummary = pd.DataFrame({'Gender': ['Male', 'Female', 'Other / Non-Disclosed'],
                             'Total Count': [MaleCount, FemaleCount, OtherCount],
                             'Percentage of Players': [MalePercent, FemalePercent, OtherPercent]})
GenderSummary.set_index('Gender')
GenderSummary['Percentage of Players'] = GenderSummary['Percentage of Players'].map('{:.2f}%'.format)

GenderSummary
```
